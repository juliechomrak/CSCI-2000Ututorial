```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%pylab inline

z = 0
file = open ('Source_data_for_CFR_vaccine_map_abridged.csv')
disease = np.array([])
x = np.array([])
y = np.array([])
cases = np.array([])
z = 0

for line in file:
    row = line.split(',')

    for i in range(0,4):
        if (row[i] == '\n'):
            row[i] = "1"
        if (row[i] == ''):
            row[i] = "0"
        row[i] = row[i].replace('\n','')
        row[i] = row[i].replace('"','')
        row[i] = row[i].replace('unknown','1')
        row[i] = row[i].replace('Unknown','1')
        row[i] = row[i].replace('','')
        
    
    disease = np.append(disease,(row[0]))
    x = np.append(x,float(row[1]))
    y = np.append (y,float(row[2]))
    cases = np.append(cases,float(row[3]))
    
    
fig = plt.figure(figsize=(14,10))
ax = plt.subplot(1,1,1)

earth = Basemap(projection='mill')

earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

for element in x:
    if (disease[z] == 'Measles'):
        ax.scatter(x[z],y[z],s = cases[z],c = 'Red')
    z = z + 1
z = 0

for element in x:
    if (disease[z] == 'Rubella'):
        ax.scatter(x[z],y[z],s = cases[z],c = 'Magenta')
    z = z + 1
z = 0

for element in x:
    if (disease[z] == 'Mumps'):
        ax.scatter(x[z],y[z],s = cases[z],c = 'Green')
    z = z + 1
z = 0

for element in x:
    if (disease[z] == 'Polio'):
        ax.scatter(x[z],y[z],s = cases[z],c = 'Blue')
    z = z + 1
z = 0


plt.show()
```
