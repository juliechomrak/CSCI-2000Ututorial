```python
#Julie Chomrak 100555212
%matplotlib inline
import matplotlib.pyplot as plt
#sets the number of medals for each country in order
x = [46,38,29,24,13,11,11,8,8,7,107]
#sts the county name in order
y = ["USA","CHN","GBR","RUS","KOR","GER","FRA","ITA","HUN","AUS","OTHER"]
#plots the pie chart using the x and y arrays
plt.pie(x, explode=None, labels=y,
    colors=('b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'),
    autopct=None, pctdistance=0.6, shadow=False,
    labeldistance=1.1, startangle=None, radius=None,
    counterclock=True, wedgeprops=None, textprops=None)
#creates a title for the pie chart
plt.title("2012 Olympic gold medals (100555212)", fontsize = 16) 

plt.show()
```
