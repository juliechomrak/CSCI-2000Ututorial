```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#julie Chomrak 100555212

#opens xy_points.dat in read only mode
file = open('xy_points.dat','r')
#switch variable to switch between x and y arrays
z = 0
#defining the 3 needed arrays
x = np.array([])
y = np.array([])
r = np.array([])
#for loop with nested ifs to load the coordinates into the arrays
#converting words to float since they cannot be understoof in any other form
#appends the array with the new word adding it to a new cell at the end of the array
for line in file:
    for word in line.split():
        if (z == 1):
            z -= 1
            float(word)
            x = np.append(x,word)
        elif (z == 0): 
            z += 1
            float(word)
            y = np.append(y,word)
#plots all points using the arrays x and y
plt.scatter(x, y, c='red')
#sets x and y axis maxes and mins
plt.xlim(-10, 10)
plt.ylim(-10, 10)
#creates the title
plt.title("Highlighted Points (100555212)")
#plots points based on the equations from part c
plt.scatter(x[[-5<=y<=0],[x<-2]], y[[-5<=y<=0],[x<-2]], c='blue')
#for loop to fill the r array
for i in range (0,999):
    #produces an output for each x and y coordinate pair using the equation given in part d
    f = ((((float(x[i])-0)**2)+(float(y[i])-5)**2)**(1/2))
    #appends output to r array
    r = np.append(r, f)
    #plots points who's coordinates are within 5units of the origine point (0,5)
plt.scatter(x[r<=5],y[r<=5], c='green')
```
